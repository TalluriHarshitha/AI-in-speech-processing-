**Segment 1**

```python
from IPython.display import Audio
y, sr = librosa.load(filename) 
sp= librosa.effects.split(y, top_db=30)
print(sp)
```

```python
segments=[y[start:end] for start,end in sp]
print(len(segments))
```

```python
audio_seg1=Audio(segments[0],rate=sr)
print('segment 1')
display(audio_seg1)
```

```python
audio_seg2=Audio(segments[1],rate=sr)
print('segment 2')
display(audio_seg2)
```

```python
audio_seg3=Audio(segments[2],rate=sr)
print('segment 3')
display(audio_seg3)
```

```python
audio_seg4=Audio(segments[3],rate=sr)
print('segment 4')
display(audio_seg4)
```
